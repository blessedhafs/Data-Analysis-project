## HYPOTHESIS TESTING - CUSTOMER SATISFACTION 

You're working as a customer experience analyst at a mid-sized telecommunications company. 
The company recently implemented a new customer service improvement strategy, which includes 
a streamlined service request process, additional training for service agents, and clearer 
communication templates. 

Management believes this strategy has improved customer satisfaction 
across several service dimensions such as service speed, communication clarity, and after-sales 
support. 

To objectively assess whether the new strategy has made a significant difference, you 
decide to compare customer satisfaction ratings before and after the strategy was implemented. 


You've conducted a survey using a 10-item customer satisfaction scale, with responses rated from 
1 (Very Dissatisfied) to 5 (Very Satisfied). Each customer answered the same set of questions 
before and after the strategy was introduced. 


# Research Question 

Has the customer service improvement strategy significantly increased overall customer 
satisfaction scores? 


```python
# import necessary libraries and load dataset
import pandas as pd
import scipy.stats as st

cust_sat_bef = pd.read_excel(r"C:\Users\USER\Desktop\DATA ANALYSIS CLASS\VSCODE\DATASET\PROJECT\customer_satisfaction_before_after (1).xlsx",
                              sheet_name= "Before")
# sum up each value
cust_sat_bef["Total"] = cust_sat_bef.iloc[:,1:].sum(axis = 1)
cust_sat_bef.head()
```

```python
cust_sat_aft = pd.read_excel(r"C:\Users\USER\Desktop\DATA ANALYSIS CLASS\VSCODE\DATASET\PROJECT\customer_satisfaction_before_after (1).xlsx", 
                             sheet_name = "After")
# sum up each value
cust_sat_aft["Total"] = cust_sat_aft.iloc[:,1:].sum(axis = 1)
cust_sat_aft.head()
```


# Research Hypothesis 

H₀: There is no difference in mean satisfaction before and after the strategy. 

H₁: Mean satisfaction after the strategy is higher than before. 

```python
result = st.ttest_rel(cust_sat_aft["Total"], cust_sat_bef["Total"])
print(f"Statistic: {result[0]:.2f}\nPvalue: {result[1]}")
result
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame({"Before": cust_sat_bef["Total"],
             "After": cust_sat_aft["Total"]})
plt.figure(figsize=(10,5))
sns.boxplot(data)
plt.title("Boxplot of Customer satisfaction survey")
```

## Analysis Result

The result of the paired t-test analysis carried out on the customer satisfaction survey, as shown above, reveals that the p-value is less than 0.05. Hence, we reject the null hypothesis. This indicates that there is a significant difference between the mean customer satisfaction before and after the new strategy was adopted. The plot above shows that satisfaction after the strategy is higher than before.
