## EXPLORING NYC PUBLIC SCHOOL TEST RESULT SCORES

Every year, American high school students take SATs, which are standardized tests intended to 
measure literacy, numeracy, and writing skills. There are three sections - reading, math, and 
writing, each with a maximum score of 800 points. These tests are extremely important for students 
and colleges, as they play a pivotal role in the admissions process. 
Analyzing the performance of schools is important for a variety of stakeholders, including policy 
and education professionals, researchers, government, and even parents considering which school 
their children should attend. 

You have been provided with a dataset called schools.csv. You have been tasked with answering 
three key questions about New York City (NYC) public school SAT performance. 

```python
import pandas as pd
sch = pd.read_csv(r"C:\Users\USER\Desktop\DATA ANALYSIS CLASS\VSCODE\DATASET\PROJECT\schools (3).csv")
sch
```

1. Which NYC schools have the best math results? 

• The best math results are at least 80% of the *maximum possible score of 800* for 
math. 

• Save your results in a pandas DataFrame called best_math_schools, including 
"school_name" and "average_math" columns, sorted by "average_math" in descending 
order. 


```python
sch["borough"].unique()
sch["average_math"].max()
cut_off = 0.8 * 800
best_math_schools = sch.query("average_math >= 0.8 * 800").sort_values(by = ["average_math"], ascending=False).reset_index()
best_math_schools
```

2. What are the top 10 performing schools based on the combined SAT scores? 

• Save your results as a pandas DataFrame called top_10_schools containing the 
"school_name" and a new column named "total_SAT", with results ordered by 
"total_SAT" in descending order ("total_SAT" being the sum of math, reading, and 
writing scores). 

```python
sch["Total_SAT"] = sch.iloc[:, -4:-1].sum(axis = 1)
top_10_schools = sch.nlargest(10, "Total_SAT").reset_index()
top_10_schools
```

3. Which single borough has the largest standard deviation in the combined SAT score? 

• Save your results as a pandas DataFrame called largest_std_dev. 

• The DataFrame should contain one row, with: 

• "borough" - the name of the NYC borough with the largest standard deviation 
of "total_SAT". 

• "num_schools" - the number of schools in the borough. 

• "average_SAT" - the mean of "total_SAT". 

• "std_SAT" - the standard deviation of "total_SAT". 

• Round all numeric values to two decimal places. 

```python
std_dev_tab = sch.groupby("borough")[["Total_SAT", "school_name"]].agg({"school_name": "count", "Total_SAT":["mean", "std"]}).round(2)
max_std = std_dev_tab['Total_SAT','std'].max() 
largest_std_dev = std_dev_tab[std_dev_tab['Total_SAT','std'] >= max_std]
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
largest_std_dev
```
